```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def open_file(path = "avP-R11-S10I5R5.dat"):
    data = []
    with open(path) as f:
        for l in f.readlines():
            line = (l.split())
            if(line[0][0]!='#'):
              line = list(map(float, line))
              data.append(line)
    data = np.array(data)
    return data
```

```python
avP11 = open_file(path = "avP-R11-S10I5R5.dat")
avP12 = open_file(path = "avP-R12-S10I5R5.dat")
avP21 = open_file(path = "avP-R21-S10I5R5.dat")
avP22 = open_file(path = "avP-R22-S10I5R5.dat")
avQ11 = open_file(path = "avQ-R11-S10I5R5.dat")
avQ12 = open_file(path = "avQ-R12-S10I5R5.dat")
avQ21 = open_file(path = "avQ-R21-S10I5R5.dat")
avQ22 = open_file(path = "avQ-R22-S10I5R5.dat")
```

```python
blue = sns.color_palette("Blues_d", n_colors=3)
red=sns.color_palette("Reds_d", n_colors=3)
orange=sns.color_palette("Oranges_d", n_colors=3)
green=sns.color_palette("Greens_d", n_colors=3)
sns.palplot(blue), sns.palplot(red), sns.palplot(orange), sns.palplot(green)
```

```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

sns.set(context="paper", style="white")
sns.set_style("ticks", {"xtick.major.size": 10, "ytick.major.size": 10})
fig, axs = plt.subplots(ncols=3, nrows=1, figsize=(15,5), squeeze=False)

axis_font = {'fontname':'Arial', 'size':'20'}

kwargs = dict(histtype='stepfilled', alpha=0.7, normed=True, bins=40)
              
#1,1 plot
ax1 = axs[0][0]
ax1.set_title("(a)",fontsize='x-large',loc='right',y=1,fontweight ="bold")
#ax1.set_yscale("log")
x1=avP11[:,1]
x2=avP12[:,1]
x3=avP21[:,1]
x4=avP22[:,1]
ax1.hist(x1, **kwargs)
ax1.hist(x2, **kwargs)
ax1.hist(x3, **kwargs)
ax1.hist(x4, **kwargs)
ax1.set_xlabel('$p_R^{\mu}$',**axis_font)
ax1.legend(["R11","R12","R21","R22"], loc="best", fontsize="x-large")
ax1.tick_params(axis='both',labelsize=16)


#1,2 plot
ax2 = axs[0][1]
ax2.set_title("(b)",fontsize='x-large',loc='right',y=1,fontweight ="bold")
#ax2.set_yscale("log")
x1=avP11[:,3]
x2=avP12[:,3]
x3=avP21[:,3]
x4=avP22[:,3]
ax2.hist(x1, **kwargs)
ax2.hist(x2, **kwargs)
ax2.hist(x3, **kwargs)
ax2.hist(x4, **kwargs)
ax2.set_xlabel('$q_R^{\mu}$',**axis_font)
ax2.legend(["R11","R12","R21","R22"], loc="best", fontsize="x-large")
ax2.tick_params(axis='both',labelsize=16)


#1,3 plot
ax3 = axs[0][2]
ax3.set_xlim([0,0.004])
ax3.set_title("(c)",fontsize='x-large',loc='right',y=1,fontweight ="bold")
#ax2.set_yscale("log")
x1=avQ11[:,1]
x2=avQ12[:,1]
x3=avQ21[:,1]
x4=avQ22[:,1]
ax3.hist(x1, **kwargs)
ax3.hist(x2, **kwargs)
ax3.hist(x3, **kwargs)
ax3.hist(x4, **kwargs)
ax3.set_xlabel('$q_{R}^{\mu\mu}$',**axis_font)
ax3.legend(["R11","R12","R21","R22"], loc="left", fontsize="x-large")
ax3.tick_params(axis='both',labelsize=16)



########################################################################



fig.tight_layout()
plt.savefig("fig4.pdf", dpi=300)
plt.show()
```
