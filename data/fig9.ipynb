```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def open_file(path = "avE-R11-S6R1.dat"):
    data = []
    with open(path) as f:
        for l in f.readlines():
            line = (l.split())
            if(len(line)>0):
              if(line[0][0]!='#'):
                line = list(map(float, line))
                data.append(line)
    data = np.array(data)
    return data
```

```python
E11S6R1 = open_file(path = "avE-R11-S6R1.dat")
E11S8R1 = open_file(path = "avE-R11-S8R1.dat")
E11S10R1 = open_file(path = "avE-R11-S10R1.dat")
E21S6R1 = open_file(path = "avE-R21-S6R1.dat")
E21S8R1 = open_file(path = "avE-R21-S8R1.dat")
E21S10R1 = open_file(path = "avE-R21-S10R1.dat")
```

```python
blue = sns.color_palette("Blues_d", n_colors=3)
red=sns.color_palette("Reds_d", n_colors=3)
orange=sns.color_palette("Oranges_d", n_colors=3)
green=sns.color_palette("Greens_d", n_colors=3)
sns.palplot(blue), sns.palplot(red), sns.palplot(orange), sns.palplot(green)
```

```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

sns.set(context="paper", style="white")
sns.set_style("ticks", {"xtick.major.size": 20, "ytick.major.size": 20})
fig, axs = plt.subplots(ncols=2, nrows=1, figsize=(10,5), squeeze=False)

axis_font = {'fontname':'Arial', 'size':'20'}


x0=np.arange(14)
y0=[0.5]*14

#1,1 plot
ax1 = axs[0][0]
ax1.set_ylim([0,1])
ax1.set_title("(a) R11",fontsize="x-large",loc="right",y=1,fontweight ="bold")
#ax1.set_xscale("log")
x1=E11S6R1[:,0]
y1=E11S6R1[:,3]
x2=E11S8R1[:,0]
y2=E11S8R1[:,3]
x3=E11S10R1[:,0]
y3=E11S10R1[:,3]
ax1.plot(x0, y0, ':', ms=8, color='k', linewidth=3, label="_nolegend_")
ax1.plot(x1, y1, '-', ms=8, color=red[0], linewidth=3, label="$N_S=6$")
ax1.plot(x2, y2, '-', ms=8, color=blue[0], linewidth=3, label="$N_S=8$")
ax1.plot(x3, y3, '-', ms=8, color=orange[1], linewidth=3, label="$N_S=10$")
# Make the y-axis label, ticks and tick labels match the line color.
ax1.set_xlabel('$P$',**axis_font)
ax1.set_ylabel(r'$\langle P_s/P \rangle$',**axis_font)
ax1.legend(loc="lower left", fontsize="large")
ax1.tick_params(axis='both',labelsize=16)


#2,1 plot
ax2 = axs[0][1]
ax2.set_ylim([0,1])
ax2.set_title("(b) R21",fontsize="x-large",loc="right",y=1,fontweight ="bold")
#ax2.set_xscale("log")
x1=E21S6R1[:,0]
y1=E21S6R1[:,3]
x2=E21S8R1[:,0]
y2=E21S8R1[:,3]
x3=E21S10R1[:,0]
y3=E21S10R1[:,3]
ax2.plot(x0, y0, ':', ms=8, color='k', linewidth=3, label="_nolegend_")
ax2.plot(x1, y1, '-', ms=8, color=red[0], linewidth=3, label="$N_S=6$")
ax2.plot(x2, y2, '-', ms=8, color=blue[0], linewidth=3, label="$N_S=8$")
ax2.plot(x3, y3, '-', ms=8, color=orange[1], linewidth=3, label="$N_S=10$")
# Make the y-axis label, ticks and tick labels match the line color.
ax2.set_xlabel('$P$',**axis_font)
ax2.set_ylabel(r'$\langle P_s/P \rangle$',**axis_font)
ax2.legend(loc="lower left", fontsize="large")
ax2.tick_params(axis='both',labelsize=16)


########################################################################



fig.tight_layout()
plt.savefig("fig9.pdf", dpi=300)
plt.show()
```
