```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def open_file(path = "avP-R11-S10R5.dat"):
    data = []
    with open(path) as f:
        for l in f.readlines():
            line = (l.split())
            if(line[0][0]!='#'):
              line = list(map(float, line))
              data.append(line)
    data = np.array(data)
    return data
```

```python
avP11 = open_file(path = "avP-R11-S10R5.dat")
avP12 = open_file(path = "avP-R12-S10R5.dat")
avP21 = open_file(path = "avP-R21-S10R5.dat")
avP22 = open_file(path = "avP-R22-S10R5.dat")
avQ11 = open_file(path = "avQ-R11-S10R5.dat")
avQ12 = open_file(path = "avQ-R12-S10R5.dat")
avQ21 = open_file(path = "avQ-R21-S10R5.dat")
avQ22 = open_file(path = "avQ-R22-S10R5.dat")
avPh11 = open_file(path = "avP-R11-S10I5R5.dat")
avPh12 = open_file(path = "avP-R12-S10I5R5.dat")
avPh21 = open_file(path = "avP-R21-S10I5R5.dat")
avPh22 = open_file(path = "avP-R22-S10I5R5.dat")
avQh11 = open_file(path = "avQ-R11-S10I5R5.dat")
avQh12 = open_file(path = "avQ-R12-S10I5R5.dat")
avQh21 = open_file(path = "avQ-R21-S10I5R5.dat")
avQh22 = open_file(path = "avQ-R22-S10I5R5.dat")
```

```python
blue = sns.color_palette("Blues_d", n_colors=3)
red=sns.color_palette("Reds_d", n_colors=3)
orange=sns.color_palette("Oranges_d", n_colors=3)
green=sns.color_palette("Greens_d", n_colors=3)
sns.palplot(blue), sns.palplot(red), sns.palplot(orange), sns.palplot(green)
```

```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from seaborn import kdeplot 

sns.set(context="paper", style="white")
sns.set_style("ticks", {"xtick.major.size": 10, "ytick.major.size": 10})
fig, axs = plt.subplots(ncols=3, nrows=2, figsize=(15,10), squeeze=False)

axis_font = {'fontname':'Arial', 'size':'20'}

kwargs = dict(histtype='stepfilled', alpha=0.7, normed=True, bins=40)
              
#1,1 plot
ax1 = axs[0][0]
ax1.set_title("(a1)",fontsize='x-large',loc='right',y=1,fontweight ="bold")
#ax1.set_yscale("log")
x1=avP11[:,1]
x2=avP12[:,1]
x3=avP21[:,1]
x4=avP22[:,1]
sns.kdeplot(x1, ax=ax1, linewidth=3, label="R11")
sns.kdeplot(x2, ax=ax1, linewidth=3, label="R12")
sns.kdeplot(x3, ax=ax1, linewidth=3, label="R21")
sns.kdeplot(x4, ax=ax1, linewidth=3, label="R22")
ax1.set_xlabel('$p_R^{\mu}$',**axis_font)
ax1.legend(["R11","R12","R21","R22"], loc="best", fontsize="large")
ax1.tick_params(axis='both',labelsize=16)


#1,2 plot
ax2 = axs[0][1]
ax2.set_title("(b1)",fontsize='x-large',loc='right',y=1,fontweight ="bold")
#ax2.set_yscale("log")
x1=avP11[:,3]
x2=avP12[:,3]
x3=avP21[:,3]
x4=avP22[:,3]
sns.kdeplot(x1, ax=ax2, linewidth=3, label="R11")
sns.kdeplot(x2, ax=ax2, linewidth=3, label="R12")
sns.kdeplot(x3, ax=ax2, linewidth=3, label="R21")
sns.kdeplot(x4, ax=ax2, linewidth=3, label="R22")
ax2.set_xlabel('$q_R^{\mu}$',**axis_font)
ax2.legend(["R11","R12","R21","R22"], loc="best", fontsize="large")
ax2.tick_params(axis='both',labelsize=16)


#1,3 plot
ax3 = axs[0][2]
ax3.set_xlim([0,0.00065])
ax3.set_title("(c1)",fontsize='x-large',loc='right',y=1,fontweight ="bold")
#ax2.set_yscale("log")
x1=avQ11[:,1]
x2=avQ12[:,1]
x3=avQ21[:,1]
x4=avQ22[:,1]
sns.kdeplot(x1, ax=ax3, linewidth=3, label="R11")
sns.kdeplot(x2, ax=ax3, linewidth=3, label="R12")
sns.kdeplot(x3, ax=ax3, linewidth=3, label="R21")
sns.kdeplot(x4, ax=ax3, linewidth=3, label="R22")
ax3.set_xlabel('$q_R^{\mu\mu}$',**axis_font)
ax3.legend(["R11","R12","R21","R22"], loc="best", fontsize="large")
ax3.tick_params(axis='both',labelsize=16)


########################################################################

#1,1 plot
ax4 = axs[1][0]
ax4.set_title("(a2)",fontsize='x-large',loc='right',y=1,fontweight ="bold")
#ax4.set_yscale("log")
x1=avPh11[:,1]
x2=avPh12[:,1]
x3=avPh21[:,1]
x4=avPh22[:,1]
sns.kdeplot(x1, ax=ax4, linewidth=3, label="R11")
sns.kdeplot(x2, ax=ax4, linewidth=3, label="R12")
sns.kdeplot(x3, ax=ax4, linewidth=3, label="R21")
sns.kdeplot(x4, ax=ax4, linewidth=3, label="R22")
ax4.set_xlabel('$p_R^{\mu}$',**axis_font)
ax4.legend(["R11","R12","R21","R22"], loc="best", fontsize="large")
ax4.tick_params(axis='both',labelsize=16)


#1,2 plot
ax5 = axs[1][1]
ax5.set_title("(b2)",fontsize='x-large',loc='right',y=1,fontweight ="bold")
#ax5.set_yscale("log")
x1=avPh11[:,3]
x2=avPh12[:,3]
x3=avPh21[:,3]
x4=avPh22[:,3]
sns.kdeplot(x1, ax=ax5, linewidth=3, label="R11")
sns.kdeplot(x2, ax=ax5, linewidth=3, label="R12")
sns.kdeplot(x3, ax=ax5, linewidth=3, label="R21")
sns.kdeplot(x4, ax=ax5, linewidth=3, label="R22")
ax5.set_xlabel('$q_R^{\mu}$',**axis_font)
ax5.legend(["R11","R12","R21","R22"], loc="best", fontsize="large")
ax5.tick_params(axis='both',labelsize=16)


#1,3 plot
ax6 = axs[1][2]
ax6.set_xlim([0,0.0035])
ax6.set_title("(c2)",fontsize='x-large',loc='right',y=1,fontweight ="bold")
#ax6.set_yscale("log")
x1=avQh11[:,1]
x2=avQh12[:,1]
x3=avQh21[:,1]
x4=avQh22[:,1]
sns.kdeplot(x1, ax=ax6, linewidth=3, label="R11")
sns.kdeplot(x2, ax=ax6, linewidth=3, label="R12")
sns.kdeplot(x3, ax=ax6, linewidth=3, label="R21")
sns.kdeplot(x4, ax=ax6, linewidth=3, label="R22")
ax6.set_xlabel('$q_R^{\mu\mu}$',**axis_font)
ax6.legend(["R11","R12","R21","R22"], loc="best", fontsize="large")
ax6.tick_params(axis='both',labelsize=16)


########################################################################


fig.tight_layout()
plt.savefig("fig2.pdf", dpi=300)
plt.show()
```
