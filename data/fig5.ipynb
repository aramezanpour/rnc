```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def open_file(path = "LS-R11-N10.dat"):
    data = []
    with open(path) as f:
        for l in f.readlines():
            line = (l.split())
            if(len(line)>0):
              if(line[0][0]!='#'):
                line = list(map(float, line))
                data.append(line)
    data = np.array(data)
    return data
```

```python
LS11N10 = open_file(path = "LS-R11-N10.dat")
LS11N20 = open_file(path = "LS-R11-N20.dat")
LS11N40 = open_file(path = "LS-R11-N40.dat")
LS21N10 = open_file(path = "LS-R21-N10.dat")
LS21N20 = open_file(path = "LS-R21-N20.dat")
LS21N40 = open_file(path = "LS-R21-N40.dat")
LS21MN10 = open_file(path = "LS-R21-MN10.dat")
LS21MN20 = open_file(path = "LS-R21-MN20.dat")
LS21MN40 = open_file(path = "LS-R21-MN40.dat")
```

```python
blue = sns.color_palette("Blues_d", n_colors=3)
red=sns.color_palette("Reds_d", n_colors=3)
orange=sns.color_palette("Oranges_d", n_colors=3)
green=sns.color_palette("Greens_d", n_colors=3)
sns.palplot(blue), sns.palplot(red), sns.palplot(orange), sns.palplot(green)
```

```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

sns.set(context="paper", style="white")
sns.set_style("ticks", {"xtick.major.size": 20, "ytick.major.size": 20})
fig, axs = plt.subplots(ncols=3, nrows=2, figsize=(15,10), squeeze=False)

axis_font = {'fontname':'Arial', 'size':'20'}
              
#1,1 plot
ax1 = axs[0][0]
ax1.set_title("(a1) R11",fontsize="x-large",loc='right', y=1,fontweight ="bold")
ax1.set_xlim([0.5,12.5])
#ax1.set_yscale("log")
x1=LS11N10[:,0]
y1=LS11N10[:,1]
x2=LS11N20[:,0]
y2=LS11N20[:,1]
x3=LS11N40[:,0]
y3=LS11N40[:,1]
ax1.plot(x1, y1, '-o', ms=8, color=red[2], label="$N_S=10$")
ax1.plot(x2, y2, '-s', ms=8, color=blue[2], label="$N_S=20$")
ax1.plot(x3, y3, '-d', ms=8, color=green[0], label="$N_S=40$")
# Make the y-axis label, ticks and tick labels match the line color.
ax1.set_xlabel(r'$P$',**axis_font)
ax1.set_ylabel('Probability',**axis_font)
ax1.legend(loc="best", fontsize="large")
ax1.tick_params(axis='both',labelsize=16)

            
#1,2 plot
ax2 = axs[0][1]
ax2.set_title("(b1) R21",fontsize="x-large",loc='right',y=1,fontweight ="bold")
ax2.set_xlim([0.5,20.5])
#ax2.set_yscale("log")
x1=LS21N10[:,0]
y1=LS21N10[:,1]
x2=LS21N20[:,0]
y2=LS21N20[:,1]
x3=LS21N40[:,0]
y3=LS21N40[:,1]
ax2.plot(x1, y1, '-o', ms=8, color=red[2], label="$N_S=10$")
ax2.plot(x2, y2, '-s', ms=8, color=blue[2], label="$N_S=20$")
ax2.plot(x3, y3, '-d', ms=8, color=green[0], label="$N_S=40$")
# Make the y-axis label, ticks and tick labels match the line color.
ax2.set_xlabel(r'$P$',**axis_font)
ax2.set_ylabel('',**axis_font)
ax2.legend(loc="best", fontsize="large")
ax2.tick_params(axis='both',labelsize=16)


            
#1,2 plot
ax2 = axs[0][2]
ax2.set_title("(c1) R21 ($M=N_S$)",fontsize="x-large",loc='right',y=1,fontweight ="bold")
ax2.set_xlim([0.5,20.5])
#ax2.set_yscale("log")
x1=LS21MN10[:,0]
y1=LS21MN10[:,1]
x2=LS21MN20[:,0]
y2=LS21MN20[:,1]
x3=LS21MN40[:,0]
y3=LS21MN40[:,1]
ax2.plot(x1, y1, '-o', ms=8, color=red[2], label="$N_S=10$")
ax2.plot(x2, y2, '-s', ms=8, color=blue[2], label="$N_S=20$")
ax2.plot(x3, y3, '-d', ms=8, color=green[0], label="$N_S=40$")
# Make the y-axis label, ticks and tick labels match the line color.
ax2.set_xlabel(r'$P$',**axis_font)
ax2.set_ylabel('',**axis_font)
ax2.legend(loc="best", fontsize="large")
ax2.tick_params(axis='both',labelsize=16)


########################################################################

#1,1 plot
ax1 = axs[1][0]
ax1.set_title("(a2) R11",fontsize="x-large",loc='right', y=1,fontweight ="bold")
ax1.set_xlim([0,6])
#ax1.set_yscale("log")
x1=LS11N10[:,0]/pow(10,0.25)
y1=LS11N10[:,1]
x2=LS11N20[:,0]/pow(20,0.25)
y2=LS11N20[:,1]
x3=LS11N40[:,0]/pow(40,0.25)
y3=LS11N40[:,1]
ax1.plot(x1, y1, '-', ms=8, color=red[2], linewidth=3, label="$N_S=10$")
ax1.plot(x2, y2, '--', ms=8, color=blue[2], linewidth=3, label="$N_S=20$")
ax1.plot(x3, y3, '-.', ms=8, color=green[0], linewidth=3, label="$N_S=40$")
# Make the y-axis label, ticks and tick labels match the line color.
ax1.set_xlabel(r'$P/N_S^{\nu}$',**axis_font)
ax1.set_ylabel('Probability',**axis_font)
ax1.legend(loc="best", fontsize="large")
ax1.tick_params(axis='both',labelsize=16)

            
#1,2 plot
ax2 = axs[1][1]
ax2.set_title("(b2) R21",fontsize="x-large",loc='right',y=1,fontweight ="bold")
ax2.set_xlim([0,4])
#ax2.set_yscale("log")
x1=LS21N10[:,0]/pow(10,0.5)
y1=LS21N10[:,1]
x2=LS21N20[:,0]/pow(20,0.5)
y2=LS21N20[:,1]
x3=LS21N40[:,0]/pow(40,0.5)
y3=LS21N40[:,1]
ax2.plot(x1, y1, '-', ms=8, color=red[2], linewidth=3, label="$N_S=10$")
ax2.plot(x2, y2, '--', ms=8, color=blue[2], linewidth=3, label="$N_S=20$")
ax2.plot(x3, y3, '-.', ms=8, color=green[0], linewidth=3, label="$N_S=40$")
# Make the y-axis label, ticks and tick labels match the line color.
ax2.set_xlabel(r'$P/N_S^{\nu}$',**axis_font)
ax2.set_ylabel('',**axis_font)
ax2.legend(loc="best", fontsize="large")
ax2.tick_params(axis='both',labelsize=16)


            
#1,2 plot
ax2 = axs[1][2]
ax2.set_title("(c2) R21 ($M=N_S$)",fontsize="x-large",loc='right',y=1,fontweight ="bold")
ax2.set_xlim([0,4])
#ax2.set_yscale("log")
x1=LS21MN10[:,0]/pow(10,0.5)
y1=LS21MN10[:,1]
x2=LS21MN20[:,0]/pow(20,0.5)
y2=LS21MN20[:,1]
x3=LS21MN40[:,0]/pow(40,0.5)
y3=LS21MN40[:,1]
ax2.plot(x1, y1, '-', ms=8, color=red[2], linewidth=3, label="$N_S=10$")
ax2.plot(x2, y2, '--', ms=8, color=blue[2], linewidth=3, label="$N_S=20$")
ax2.plot(x3, y3, '-.', ms=8, color=green[0], linewidth=3, label="$N_S=40$")
# Make the y-axis label, ticks and tick labels match the line color.
ax2.set_xlabel(r'$P/N_S^{\nu}$',**axis_font)
ax2.set_ylabel('',**axis_font)
ax2.legend(loc="best", fontsize="large")
ax2.tick_params(axis='both',labelsize=16)


########################################################################



fig.tight_layout()
plt.savefig("fig5.pdf", dpi=300)
plt.show()
```
