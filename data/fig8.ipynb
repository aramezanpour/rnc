```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def open_file(path = "avE-R11-S8I4R1-P1-r.dat"):
    data = []
    with open(path) as f:
        for l in f.readlines():
            line = (l.split())
            if(len(line)>0):
              if(line[0][0]!='#'):
                line = list(map(float, line))
                data.append(line)
    data = np.array(data)
    return data
```

```python
E11S8R1P1 = open_file(path = "avE-R11-S8I4R1-P1-r.dat")
E11S8R1P2 = open_file(path = "avE-R11-S8I4R1-P2-r.dat")
E11S8R1P3 = open_file(path = "avE-R11-S8I4R1-P3-r.dat")
E11S6R1P2 = open_file(path = "avE-R11-S6I3R1-P2-r.dat")
E11S10R1P2 = open_file(path = "avE-R11-S10I5R1-P2-r.dat")
E21S8R1P1 = open_file(path = "avE-R21-S8I4R1-P1-r.dat")
E21S8R1P2 = open_file(path = "avE-R21-S8I4R1-P2-r.dat")
E21S8R1P3 = open_file(path = "avE-R21-S8I4R1-P3-r.dat")
E21S6R1P2 = open_file(path = "avE-R21-S6I3R1-P2-r.dat")
E21S10R1P2 = open_file(path = "avE-R21-S10I5R1-P2-r.dat")
```

```python
blind = sns.color_palette("colorblind", n_colors=6)
blue = sns.color_palette("Blues_d", n_colors=3)
red=sns.color_palette("Reds_d", n_colors=3)
orange=sns.color_palette("Oranges_d", n_colors=3)
green=sns.color_palette("Greens_d", n_colors=3)
sns.palplot(blue), sns.palplot(red), sns.palplot(orange), sns.palplot(green), sns.palplot(blind)
```

```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

sns.set(context="paper", style="white")
sns.set_style("ticks", {"xtick.major.size": 20, "ytick.major.size": 20})
fig, axs = plt.subplots(ncols=2, nrows=2, figsize=(10,10), squeeze=False)

axis_font = {'fontname':'Arial', 'size':'20'}
     
x0=np.arange(2000)
y0=[0.5]*2000

#1,1 plot
ax1 = axs[0][0]
ax1.set_title(r"(a) R11, $N_S$=8",fontsize="x-large",loc="right",y=1,fontweight ="bold")
ax1.set_ylim([0,1.02])
ax1.set_xscale("log")
x1=E11S8R1P1[:,0]
y1=E11S8R1P1[:,3]
x2=E11S8R1P2[:,0]
y2=E11S8R1P2[:,3]
x3=E11S8R1P3[:,0]
y3=E11S8R1P3[:,3]
ax1.plot(x0, y0, ':', ms=8, color='k', linewidth=3, label="_nolegend_")
ax1.plot(x1, y1, '-', ms=8, color=blind[0], linewidth=3, label="$P=1$")
ax1.plot(x2, y2, '-', ms=8, color=blind[1], linewidth=3, label="$P=2$")
ax1.plot(x3, y3, '-', ms=8, color=blind[2], linewidth=3, label="$P=3$")
ax1.set_xlabel('$T$',**axis_font)
ax1.set_ylabel(r'$\langle P_s/P \rangle$',**axis_font)
ax1.legend(loc="lower right", fontsize="large")
ax1.tick_params(axis='both',labelsize=16)


#2,1 plot
ax2 = axs[1][0]
ax2.set_title("(c) R11, P=2",fontsize="x-large",loc="right",y=1,fontweight ="bold")
ax2.set_ylim([0,1])
ax2.set_xscale("log")
x1=E11S6R1P2[:,0]
y1=E11S6R1P2[:,3]
x2=E11S8R1P2[:,0]
y2=E11S8R1P2[:,3]
x3=E11S10R1P2[:,0]
y3=E11S10R1P2[:,3]
ax2.plot(x0, y0, ':', ms=8, color='k', linewidth=3, label="_nolegend_")
ax2.plot(x1, y1, '-', ms=8, color=blind[0], linewidth=3, label="$N_S=6$")
ax2.plot(x2, y2, '-', ms=8, color=blind[1], linewidth=3, label="$N_S=8$")
ax2.plot(x3, y3, '-', ms=8, color=blind[2], linewidth=3, label="$N_S=10$")
# Make the y-axis label, ticks and tick labels match the line color.
ax2.set_xlabel('$T$',**axis_font)
ax2.set_ylabel(r'$\langle P_s/P \rangle$',**axis_font)
ax2.legend(loc="lower right", fontsize="large")
ax2.tick_params(axis='both',labelsize=16)


#1,2 plot
ax3 = axs[0][1]
ax3.set_title(r"(b) R21, $N_S$=8",fontsize="x-large",loc="right",y=1,fontweight ="bold")
ax3.set_ylim([0,1.02])
ax3.set_xscale("log")
x1=E21S8R1P1[:,0]
y1=E21S8R1P1[:,3]
x2=E21S8R1P2[:,0]
y2=E21S8R1P2[:,3]
x3=E21S8R1P3[:,0]
y3=E21S8R1P3[:,3]
ax3.plot(x0, y0, ':', ms=8, color='k', linewidth=3, label="_nolegend_")
ax3.plot(x1, y1, '-', ms=8, color=blind[0], linewidth=3, label="$P=1$")
ax3.plot(x2, y2, '-', ms=8, color=blind[1], linewidth=3, label="$P=2$")
ax3.plot(x3, y3, '-', ms=8, color=blind[2], linewidth=3, label="$P=3$")
ax3.set_xlabel('$T$',**axis_font)
ax3.set_ylabel(r'$\langle P_s/P \rangle$',**axis_font)
ax3.legend(loc="lower right", fontsize="large")
ax3.tick_params(axis='both',labelsize=16)


#2,2 plot
ax4 = axs[1][1]
ax4.set_title("(d) R21, P=2",fontsize="x-large",loc="right",y=1,fontweight ="bold")
ax4.set_ylim([0,1])
ax4.set_xscale("log")
x1=E21S6R1P2[:,0]
y1=E21S6R1P2[:,3]
x2=E21S8R1P2[:,0]
y2=E21S8R1P2[:,3]
x3=E21S10R1P2[:,0]
y3=E21S10R1P2[:,3]
ax4.plot(x0, y0, ':', ms=8, color='k', linewidth=3, label="_nolegend_")
ax4.plot(x1, y1, '-', ms=8, color=blind[0], linewidth=3, label="$N_S=6$")
ax4.plot(x2, y2, '-', ms=8, color=blind[1], linewidth=3, label="$N_S=8$")
ax4.plot(x3, y3, '-', ms=8, color=blind[2], linewidth=3, label="$N_S=10$")
# Make the y-axis label, ticks and tick labels match the line color.
ax4.set_xlabel('$T$',**axis_font)
ax4.set_ylabel(r'$\langle P_s/P \rangle$',**axis_font)
ax4.legend(loc="lower right", fontsize="large")
ax4.tick_params(axis='both',labelsize=16)


########################################################################



fig.tight_layout()
plt.savefig("fig8.pdf", dpi=300)
plt.show()
```
