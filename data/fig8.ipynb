```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def open_file(path = "Dt-R11-S6R1-400t400-P1.dat"):
    data = []
    with open(path) as f:
        for l in f.readlines():
            line = (l.split())
            if(len(line)>0):
              if(line[0][0]!='#'):
                line = list(map(float, line))
                data.append(line)
    data = np.array(data)
    return data
```

```python
D11S6t44 = open_file(path = "Dt-R11-S6R1-400t400-P1.dat")
D11S6t14 = open_file(path = "Dt-R11-S6R1-100t400-P1.dat")
D11S6t41 = open_file(path = "Dt-R11-S6R1-400t100-P1.dat")
D11S8t44 = open_file(path = "Dt-R11-S8R1-400t400-P1.dat")
D11S8t14 = open_file(path = "Dt-R11-S8R1-100t400-P1.dat")
D11S8t41 = open_file(path = "Dt-R11-S8R1-400t100-P1.dat")
D11S10t44 = open_file(path = "Dt-R11-S10R1-400t400-P1.dat")
D11S10t14 = open_file(path = "Dt-R11-S10R1-100t400-P1.dat")
D11S10t41 = open_file(path = "Dt-R11-S10R1-400t100-P1.dat")
D21S6t44 = open_file(path = "Dt-R21-S6R1-400t400-P1.dat")
D21S6t14 = open_file(path = "Dt-R21-S6R1-100t400-P1.dat")
D21S6t41 = open_file(path = "Dt-R21-S6R1-400t100-P1.dat")
D21S8t44 = open_file(path = "Dt-R21-S8R1-400t400-P1.dat")
D21S8t14 = open_file(path = "Dt-R21-S8R1-100t400-P1.dat")
D21S8t41 = open_file(path = "Dt-R21-S8R1-400t100-P1.dat")
D21S10t44 = open_file(path = "Dt-R21-S10R1-400t400-P1.dat")
D21S10t14 = open_file(path = "Dt-R21-S10R1-100t400-P1.dat")
D21S10t41 = open_file(path = "Dt-R21-S10R1-400t100-P1.dat")
```

```python
blind = sns.color_palette("colorblind", n_colors=6)
blue = sns.color_palette("Blues_d", n_colors=3)
red=sns.color_palette("Reds_d", n_colors=3)
orange=sns.color_palette("Oranges_d", n_colors=3)
green=sns.color_palette("Greens_d", n_colors=3)
sns.palplot(blue), sns.palplot(red), sns.palplot(orange), sns.palplot(green)
```
